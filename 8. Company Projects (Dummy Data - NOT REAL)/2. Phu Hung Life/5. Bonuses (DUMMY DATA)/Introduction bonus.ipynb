```python
import numpy as np
import pandas as pd
import pyodbc as pyo
import sqlalchemy as sa
import numpy as np
import math
import datetime
import datetime as dt
from datetime import timedelta
#Kết nối SV_UAT (SQL_SV64)
connection_uri_UAT = (
    "mssql+pyodbc://PHLHCMHOPPBI02\PBIRS/PowerBI?driver=ODBC+Driver+17+for+SQL+Server"
)
conn = sa.create_engine(connection_uri_UAT)
```

<h3> 1. Lấy bảng

<h5>1.1. Lấy bảng AgentInfo

```python
cutoff = input("Nhập kỳ CUTOFF cuối tháng YYYYMMDD: ")
```

```python
agent_query = f"""\n
SELECT * FROM [DPO].[Main_AGENT_INFO_DA_CUTOFF]
WHERE CUTOFF = {cutoff}
"""
agent_table = pd.read_sql(agent_query, conn)
agent_table
```

<h5>1.2 Lấy bảng Commision

```python
com_query = f"""\n
SELECT * FROM [DPO].[DP_TAGENTCOMM_TEST_CUTOFF]
WHERE CUTOFF = {cutoff}
"""
com_table = pd.read_sql(com_query, conn)
com_table
```

<h3>2. Xây cấu trúc bảng 

<h5>2.1 Lấy ra quan hệ giữa người giới thiệu và được giới thiệu

```python
introduce_agent = agent_table
introduce_agent = introduce_agent[['Agent Numer of Introducer', 'Agent_Number']]
introduce_agent = introduce_agent.dropna()
introduce_agent = introduce_agent.rename(columns={'Agent Numer of Introducer':"Introducer", 'Agent_Number': "Introducee"})
introduce_agent
```

<h5>2.2 Móc lấy thông tin cụ thể của người giới thiệu:

```python
introduce_info = pd.merge(introduce_agent, agent_table, left_on = 'Introducer', right_on ='Agent_Number', how='left')
introduce_info = introduce_info[['Introducer', 'Agent_Name', 'Grade', 'Agent_Status', 'Date_Appointed', 'Introducee']].\
rename(columns={'Agent_Name':'IntroducerName', 'Grade': 'IntroducerGrade', 'Agent_Status':'IntroducerStatus', 'Date_Appointed':'IntroducerAppointed'})
introduce_info = introduce_info[introduce_info['IntroducerStatus'] != 'Terminated']
introduce_info
```

<h5>2.3 Móc lấy thông tin cụ thể người được giới thiệu:

```python
introducee_info = pd.merge(introduce_info, agent_table, left_on = 'Introducee', right_on ='Agent_Number', how ='left')
introducee_info = introducee_info[['Introducer', 'IntroducerName', 'IntroducerGrade', 'IntroducerStatus','IntroducerAppointed', 
        'Introducee', 'Agent_Name', 'Grade', 'Agent_Status', 'Date_Appointed', 'Area_Name', 'Sales_Unit_Code', 'Sales_Unit', 'CUTOFF']].\
rename(columns={'Agent_Name':'IntroduceeName', 'Grade': 'IntroduceeGrade', 'Agent_Status':'IntroduceeStatus', 'Date_Appointed':'IntroduceeAppointed',
               'Sales_Unit':"AD_Name", 'Sales_Unit_Code':'AD_Code'})
introducee_info['Appointed_Temp'] = np.where(introducee_info['IntroduceeAppointed'].dt.day > 10, 
        introducee_info['IntroduceeAppointed'] + pd.DateOffset(months=1), introducee_info['IntroduceeAppointed'])
introducee_info['Appointed_Temp'] = introducee_info['Appointed_Temp'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
introducee_info['Date_On_Set'] = pd.to_datetime(introducee_info['CUTOFF'].str[:4]+"-"+introducee_info['CUTOFF'].str[4:6]+"-"+introducee_info['CUTOFF'].str[6:])
introducee_info['Date_Temp'] = introducee_info['Date_On_Set'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
introducee_info['Length of Service(Months)'] = (introducee_info['Date_Temp'].dt.to_period('M').astype('int64') +1 ) - introducee_info['Appointed_Temp'].dt.to_period('M').astype('int64')
introducee_info = introducee_info.loc[:, ~introducee_info.columns.isin(['Date_On_Set', 'Date_Temp', 'Appointed_Temp'])]
introducee_info = introducee_info[introducee_info['IntroduceeStatus'] != 'Terminated']
introducee_info
```

<h3>3. Tính hoa hồng

<h5>3.1 Lấy ra ngày đầu tiên (thay đổi theo các ngày CUTOFF):

```python
com_t = com_table
com_t['Date_On_Set'] = pd.to_datetime(com_t['CUTOFF'].str[:4]+"-"+com_t['CUTOFF'].str[4:6]+"-"+com_t['CUTOFF'].str[6:])
if ('20230531' in com_t['CUTOFF'].values):
    com_t.loc[:,'Start_Month'] = '2023-04-30'
elif ('20220228' in com_t['CUTOFF'].values):
    com_t.loc[:,'Start_Month'] = '2022-01-28'
elif ('20220429' in com_t['CUTOFF'].values):
    com_t.loc[:,'Start_Month'] = '2022-03-31'
elif ('20220531' in com_t['CUTOFF'].values):
    com_t.loc[:,'Start_Month'] = '2022-04-30' 
elif ('20240229' in com_t['CUTOFF'].values):
    com_t.loc[:,'Start_Month'] = '2024-01-27'
else: 
    com_t['Start_Month'] = (com_t['Date_On_Set'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1))
com_t
#Lấy kỳ xét và loại ra các trạng thái HD hủy, chưa pass 21 ngày FL
com_t = com_t[(com_t['Freelook'] >21)]
com_t = com_t[~com_t['Policy Status'].isin(['CF', 'DC', 'FL', 'NT', 'WD', 'PO', 'SU', 'TR'])]
com_t
```

<h5>3.2 Tính hoa hồng theo hợp đồng 

```python
#Cộng thêm 21 ngày ACK để có kỳ xét hoa hồng chính thức
cal_com = com_t
cal_com['Valid_Date'] = np.where(cal_com['Receive Policy date'] >= cal_com['Applied Premium Date'], cal_com['Receive Policy date'], cal_com['Applied Premium Date'])
cal_com['Valid_Date'] = cal_com['Valid_Date'] + timedelta(days=22) #+22 để bảo đảm qua được 21 ngày cân nhắc
cal_com = cal_com[(cal_com['Valid_Date'] >= cal_com['Start_Month']) & (cal_com['Valid_Date'] <= cal_com['Date_On_Set'])
        & cal_com['FYC'] != 0]
cal_com = cal_com.groupby(['Commission Agent','Policy No', 'Issued Date', 'Applied Premium Date', 'Receive Policy date', 'Freelook'], dropna=False)[['FYC', 'Premium Collected']].sum().reset_index()
cal_com
```

<h3> 4. Ghép lại thông tin

```python
agent_com = pd.merge(introducee_info, cal_com, left_on = 'Introducee', right_on ='Commission Agent', how ='left')
agent_com = agent_com[agent_com['Commission Agent'].notnull()]
agent_com
```
