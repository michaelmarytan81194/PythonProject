```python
import numpy as np
import pandas as pd
import pyodbc as pyo
import sqlalchemy as sa
import numpy as np
import math
import datetime
import datetime as dt
#Kết nối SV_UAT (SQL_SV64)
connection_uri_UAT = (
    "mssql+pyodbc://PHLHCMHOPPBI02\PBIRS/PowerBI?driver=ODBC+Driver+17+for+SQL+Server"
)
conn = sa.create_engine(connection_uri_UAT)
```

<h3>1. Lấy bảng

```python
cutoff = input("Nhập kỳ CUTOFF cuối tháng YYYYMMDD: ")
```

<h5>1.1 Lấy bảng AgentInfo

```python
agent_query = f"""\n
SELECT * FROM [DPO].[Main_AGENT_INFO_DA_CUTOFF]
WHERE CUTOFF = {cutoff}
"""
agent_table = pd.read_sql(agent_query, conn)
agent_table
```

<h5>1.2 Lấy bảng PREMIUM

```python
premium_query = f"""\n
SELECT * FROM [DPO].[DP_TAGENTPREMIUM_TEST_CUTOFF]
WHERE CUTOFF = {cutoff}
"""
premium_table = pd.read_sql(premium_query, conn)
premium_table
```

<h3>2. Lấy ra ngày xét 

```python
agent_t = agent_table
```

```python
agent_t['Date_On_Set'] = pd.to_datetime(agent_t['CUTOFF'].str[:4]+"-"+agent_t['CUTOFF'].str[4:6]+"-"+agent_t['CUTOFF'].str[6:])
#Lấy ngày xét tạm:
#Rule: Join sau ngày 10 hàng tháng, tính sang tháng sau:
agent_t['Date_1st_Temp'] = np.where(agent_t['Date_Appointed'].dt.day > 10, 
        agent_t['Date_Appointed'] + pd.DateOffset(months=1), agent_t['Date_Appointed'])
agent_t['Date_1st_Temp'] = agent_t['Date_1st_Temp'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
agent_t['1st_month'] = agent_t['Date_On_Set'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1) - pd.DateOffset(months=2)
agent_t['2nd_month'] = agent_t['Date_On_Set'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1) - pd.DateOffset(months=1)
agent_t['3rd_month'] = agent_t['Date_On_Set'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
agent_t
```

<h5>2.1 Lấy ra ngày cần xét trong 3 tháng gia nhập gần nhất

```python
agent_rookie = agent_t[(agent_t['Date_1st_Temp'] >= agent_t['1st_month']) & (agent_t['Date_1st_Temp'] <= agent_t['Date_On_Set'])]
agent_rookie['Rookie_Month'] = np.where(agent_rookie['Date_1st_Temp'] == agent_rookie['3rd_month'], "1",
                            np.where(agent_rookie['Date_1st_Temp'] == agent_rookie['2nd_month'], "2",
                            "3"))
agent_rookie
```

<h5>2.2 Set Target FYP và CC:

```python
agent_target = agent_rookie
agent_target['FYP Target'] = np.where(agent_target['Rookie_Month'] == '3', 72000000,
                       np.where(agent_target['Rookie_Month'] == '2', 48000000,
                      np.where(agent_target['Rookie_Month'] == '1', 24000000, 0)))
agent_target['CC Target'] = np.where(agent_target['Rookie_Month'] == '3', 6,
                       np.where(agent_target['Rookie_Month'] == '2', 4,
                      np.where(agent_target['Rookie_Month'] == '1', 2, 0)))
agent_target                                                   
```

```python
agent_target = agent_target[['Area_Name', 'Sales_Unit_Code', 'Sales_Unit', 'Agent_Number', 'Agent_Name', 'Grade', 
         'Agent_Status', 'Date_Appointed', 'CUTOFF', 'Date_On_Set', 'Date_1st_Temp', 
        '1st_month', '2nd_month', '3rd_month', 'Rookie_Month','FYP Target', 'CC Target']]
agent_target                  
```

<h3> 3. Tính FYP 

```python
fyp_cal = premium_table
fyp_cal = fyp_cal[['Servicing Agent', 'Applied Premium Date', 'POLICY ACKNOWLED', 'Policy Status', 'FYP', 'Topup Premium']]
fyp_cal['FYP_Total'] = fyp_cal['FYP'] + fyp_cal['Topup Premium']*0.1
fyp_cal = fyp_cal.groupby(['Servicing Agent', 'Applied Premium Date', 'POLICY ACKNOWLED', 'Policy Status'])['FYP_Total'].sum().reset_index()
fyp_cal = fyp_cal[~fyp_cal['Policy Status'].isin(['CF', 'DC', 'FL',  'NT', 'WD', 'PO'])]
fyp_cal
```

<h5>3.1 Lọc các dòng FYP được tính

```python
fyp_date = pd.merge(fyp_cal, agent_target, left_on = 'Servicing Agent', right_on ='Agent_Number', how = 'left')
fyp_date = fyp_date[(fyp_date['POLICY ACKNOWLED']>=fyp_date['Date_Appointed']) & (fyp_date['POLICY ACKNOWLED']<=fyp_date['Date_On_Set'])]
#Lọc ra các TH đã pass 21 ngày FL:
fyp_date['Check_FL'] = (fyp_date['Date_On_Set'] - fyp_date['POLICY ACKNOWLED'])/pd.offsets.Day(1)
fyp_date = fyp_date[fyp_date['Check_FL'] >=21]
fyp_date
```

<h5>3.2 Tổng lấy FYP 3 tháng

```python
fyp_date = fyp_date[['Servicing Agent', 'Applied Premium Date', 'POLICY ACKNOWLED',
       'Policy Status', 'FYP_Total', 'Date_On_Set', 'Date_1st_Temp', '1st_month', '2nd_month', '3rd_month', 'Rookie_Month']]
fyp_date['Applied_Temp'] = fyp_date['POLICY ACKNOWLED'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
##BƯỚC NÀY: PHÂN RA CÁC FYP:
fyp_date['FYP_month'] = np.where(fyp_date['Applied_Temp'] <= fyp_date['1st_month'], "3",
                    np.where(fyp_date['Applied_Temp'] == fyp_date['2nd_month'], "2",
                    np.where(fyp_date['Applied_Temp'] == fyp_date['3rd_month'], "1", "")))
fyp_date = fyp_date.groupby(['Servicing Agent', 'FYP_month'])['FYP_Total'].sum().reset_index()
fyp_pivot = fyp_date.pivot(index='Servicing Agent', columns='FYP_month', values='FYP_Total').reset_index()
fyp_pivot = fyp_pivot.rename(columns={'1': 'FYP_1st', '2': 'FYP_2nd', '3': 'FYP_3rd'})
fyp_pivot
```

<h3> 4. Tính CaseCount

```python
cc_agent = premium_table
cc_agent = cc_agent[['Policy No', 'PREM TRAN NAME', 'Applied Premium Date', 'POLICY ACKNOWLED', 'Issuing Agent', 'Freelook', 'Policy Status']].drop_duplicates()
cc_agent['PREM TRAN NAME'] = cc_agent['PREM TRAN NAME'].str.strip()
cc_agent = cc_agent[(cc_agent['PREM TRAN NAME'] == 'Contract Issue') & (cc_agent['Freelook'] > 21) &
            ~cc_agent['Policy Status'].isin(['CF', 'DC', 'FL',  'NT', 'WD', 'PO'])]
cc_agent
```

<h5> 4.1 Lọc ra các ngày trong kỳ tính toán:

```python
cc_cal = pd.merge(cc_agent, agent_target, left_on = 'Issuing Agent', right_on ='Agent_Number', how = 'left')
cc_cal = cc_cal[['Policy No', 'PREM TRAN NAME', 'Applied Premium Date',
       'POLICY ACKNOWLED', 'Issuing Agent', 'Date_Appointed' ,'Freelook', 'Policy Status', 'Date_On_Set', 'Date_1st_Temp', '1st_month', '2nd_month', '3rd_month']]
cc_cal['Applied_Temp'] = cc_cal['POLICY ACKNOWLED'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
cc_cal = cc_cal[(cc_cal['POLICY ACKNOWLED'] >= cc_cal['Date_Appointed']) & (cc_cal['POLICY ACKNOWLED'] <= cc_cal['Date_On_Set'])]
cc_cal['CC_Month'] = np.where(cc_cal['Applied_Temp']<=cc_cal['1st_month'], "3",
                    np.where(cc_cal['Applied_Temp'] == cc_cal['2nd_month'], "2",
                    np.where(cc_cal['Applied_Temp'] == cc_cal['3rd_month'], "1", "")))
cc_cal = cc_cal[['Issuing Agent', 'Policy No', 'POLICY ACKNOWLED', 'CC_Month']].drop_duplicates()
cc_cal = cc_cal.groupby(['Issuing Agent', 'CC_Month'])['Policy No'].count().reset_index()
cc_cal = cc_cal.pivot(index='Issuing Agent', columns='CC_Month', values ='Policy No').reset_index()
cc_cal = cc_cal.rename(columns={'1':'CC_1st', '2': 'CC_2nd', '3':'CC_3rd'})
cc_cal
```

<h3> 5. Ghép lại thông tin

```python
agent_rookie_bonus = pd.merge(pd.merge(agent_target, fyp_pivot, left_on = 'Agent_Number', right_on ='Servicing Agent', how = 'left'), 
        cc_cal, left_on ='Agent_Number', right_on = 'Issuing Agent', how='left')
rookie_bonus = agent_rookie_bonus[['Area_Name', 'Sales_Unit_Code', 'Sales_Unit', 'Agent_Number',
       'Agent_Name', 'Grade', 'Agent_Status', 'Date_Appointed', 'CUTOFF',
       'Date_On_Set', 'Date_1st_Temp', '1st_month', '2nd_month', '3rd_month',
       'Rookie_Month', 'FYP Target', 'CC Target', 'FYP_1st', 'FYP_2nd', 'FYP_3rd', 'CC_1st', 'CC_2nd', 'CC_3rd']]
for i in ['FYP_1st', 'FYP_2nd', 'FYP_3rd', 'CC_1st', 'CC_2nd', 'CC_3rd']:
    rookie_bonus[i] = rookie_bonus[i].fillna(0)
rookie_bonus['FYP_3month_CatchUp'] = np.where(rookie_bonus['Rookie_Month'] == '3',
    rookie_bonus['FYP_1st'] + rookie_bonus['FYP_2nd'] + rookie_bonus['FYP_3rd'], 0)
rookie_bonus['FYP_2month_CatchUp'] = np.where(rookie_bonus['Rookie_Month'] == '2',
rookie_bonus['FYP_1st'] + rookie_bonus['FYP_2nd'],0)
rookie_bonus
```
