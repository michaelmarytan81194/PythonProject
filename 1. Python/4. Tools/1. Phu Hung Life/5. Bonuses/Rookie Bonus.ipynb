```python
import numpy as np
import pandas as pd
import pyodbc as pyo
import sqlalchemy as sa
import numpy as np
import math
import datetime
import datetime as dt
#Kết nối SV_UAT (SQL_SV64)
connection_uri_UAT = (
    "mssql+pyodbc://PHLHCMHOPPBI02\PBIRS/PowerBI?driver=ODBC+Driver+17+for+SQL+Server"
)
conn = sa.create_engine(connection_uri_UAT)
```

<h3>1. Lấy bảng

```python
cutoff = input("Nhập kỳ CUTOFF cuối tháng YYYYMMDD: ")
```

<h5>1.1 Lấy bảng AgentInfo

```python
agent_query = f"""\n
SELECT * FROM [DPO].[Main_AGENT_INFO_DA_CUTOFF]
WHERE CUTOFF = {cutoff}
"""
agent_table = pd.read_sql(agent_query, conn)
agent_table
```

<h5>1.2 Lấy bảng PREMIUM

```python
premium_query = f"""\n
SELECT * FROM [DPO].[DP_TAGENTPREMIUM_TEST_CUTOFF]
WHERE CUTOFF = {cutoff}
"""
premium_table = pd.read_sql(premium_query, conn)
premium_table
```

<h3>2. Lấy ra ngày xét 

```python
agent_t = agent_table
```

```python
agent_t['Date_On_Set'] = pd.to_datetime(agent_t['CUTOFF'].str[:4]+"-"+agent_t['CUTOFF'].str[4:6]+"-"+agent_t['CUTOFF'].str[6:])
#Lấy ngày xét tạm:
#Rule: Join sau ngày 10 hàng tháng, tính sang tháng sau:
agent_t['Date_1st_Temp'] = np.where(agent_t['Date_Appointed'].dt.day > 10, 
        agent_t['Date_Appointed'] + pd.DateOffset(months=1), agent_t['Date_Appointed'])
agent_t['Date_1st_Temp'] = agent_t['Date_1st_Temp'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
agent_t['1st_month'] = agent_t['Date_On_Set'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1) - pd.DateOffset(months=2)
agent_t['2nd_month'] = agent_t['Date_On_Set'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1) - pd.DateOffset(months=1)
agent_t['3rd_month'] = agent_t['Date_On_Set'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
agent_t
```

<h5>2.1 Lấy ra ngày cần xét trong 3 tháng gia nhập gần nhất

```python
agent_rookie = agent_t[(agent_t['Date_1st_Temp'] >= agent_t['1st_month']) & (agent_t['Date_1st_Temp'] <= agent_t['Date_On_Set'])]
agent_rookie['Rookie_Month'] = np.where(agent_rookie['Date_1st_Temp'] == agent_rookie['3rd_month'], "1",
                            np.where(agent_rookie['Date_1st_Temp'] == agent_rookie['2nd_month'], "2",
                            "3"))
agent_rookie
```

<h5>2.2 Set Target FYP và CC:

```python
agent_target = agent_rookie
agent_target['FYPTargetCatchup'] = np.where(agent_target['Rookie_Month'] == '3', 72000000,
                       np.where(agent_target['Rookie_Month'] == '2', 48000000,
                      np.where(agent_target['Rookie_Month'] == '1', 24000000, 0)))
agent_target['CCTargetCatchup'] = np.where(agent_target['Rookie_Month'] == '3', 6,
                       np.where(agent_target['Rookie_Month'] == '2', 4,
                      np.where(agent_target['Rookie_Month'] == '1', 2, 0)))
agent_target['FYP Target'] = 24000000
agent_target['CC Target'] = 2
```

```python
agent_target = agent_target[['Area_Name', 'Sales_Unit_Code', 'Sales_Unit', 'Agent_Number', 'Agent_Name', 'Grade', 
         'Agent_Status', 'Date_Appointed', 'CUTOFF', 'Date_On_Set', 'Date_1st_Temp', 
        '1st_month', '2nd_month', '3rd_month', 'Rookie_Month','FYP Target', 'CC Target', 'FYPTargetCatchup', 'CCTargetCatchup']]
agent_target                  
```

<h3> 3. Tính FYP 

```python
fyp_cal = premium_table
fyp_cal = fyp_cal[['Servicing Agent', 'Applied Premium Date', 'POLICY ACKNOWLED', 'Policy Status','Freelook' ,'FYP', 'Topup Premium']]
fyp_cal['FYP_Total'] = fyp_cal['FYP'] + fyp_cal['Topup Premium']*0.1
fyp_cal = fyp_cal.groupby(['Servicing Agent', 'Applied Premium Date', 'POLICY ACKNOWLED', 'Policy Status', 'Freelook'], dropna=False)['FYP_Total'].sum().reset_index()
fyp_cal = fyp_cal[~fyp_cal['Policy Status'].isin(['CF', 'DC', 'FL',  'NT', 'WD', 'PO'])]
fyp_cal
```

<h5>3.1 Lọc các dòng FYP được tính

```python
fyp_date = pd.merge(fyp_cal, agent_target, left_on = 'Servicing Agent', right_on ='Agent_Number', how = 'left')
fyp_date = fyp_date[(fyp_date['Applied Premium Date']>=fyp_date['Date_Appointed']) & (fyp_date['Applied Premium Date']<=fyp_date['Date_On_Set'])]
fyp_date = fyp_date[['Servicing Agent', 'Applied Premium Date', 'POLICY ACKNOWLED',
       'Policy Status', 'Freelook','FYP_Total', 'Date_On_Set', 'Date_1st_Temp', '1st_month', '2nd_month', '3rd_month', 'Rookie_Month']]
fyp_date['Applied_Temp'] = fyp_date['Applied Premium Date'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
fyp_date['ACK_Temp'] = fyp_date['POLICY ACKNOWLED'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
fyp_date 
```

<h5>3.2 Tổng lấy FYP theo Applied Date

```python
##BƯỚC NÀY: PHÂN RA CÁC FYP:
fyp_date['FYP_month'] = np.where(fyp_date['Applied_Temp'] == fyp_date['1st_month'], "FYP(T-2)",
                    np.where(fyp_date['Applied_Temp'] == fyp_date['2nd_month'], "FYP(T-1)",
                    np.where(fyp_date['Applied_Temp'] == fyp_date['3rd_month'], "FYP_Current(T)", "FYP(T-2)")))
fyp_date['FYP_P_month'] = np.where(fyp_date['ACK_Temp'] == fyp_date['1st_month'], "FYP_P(T-2)",
                    np.where(fyp_date['ACK_Temp'] == fyp_date['2nd_month'], "FYP_P(T-1)",
                    np.where(fyp_date['ACK_Temp'] == fyp_date['3rd_month'], "FYP_PCurrent(T)", "")))
#Note ở, bước(T-2) sau cùng, thì các TH Applied rơi vào tháng trước 1st, thì sẽ rơi vào tháng T-2 để xét
fyp_pivot = fyp_date.groupby(['Servicing Agent', 'FYP_month'], dropna=False)['FYP_Total'].sum().reset_index()
fyp_cal = fyp_pivot.pivot(index='Servicing Agent', columns='FYP_month', values='FYP_Total').reset_index()
fyp_cal[fyp_cal['Servicing Agent'] == '60073218']
```

<h5>3.3 Tổng lấy FYP cộng dồn có ngày ACK

```python
fyp_pivot_1 = fyp_date[(fyp_date['FYP_P_month']!='')].groupby(['Servicing Agent', 'FYP_P_month'], dropna=False)['FYP_Total'].sum().reset_index()
fyp_paid = fyp_pivot_1.pivot(index='Servicing Agent', columns='FYP_P_month', values='FYP_Total').reset_index()
fyp_paid[fyp_paid['Servicing Agent'] == '60073218']
```

<h5>3.4 Lấy FYP thực đạt trong tháng xét(pass 21 ngày)

```python
fyp_within_month = fyp_date
fyp_within_month['FYP_Monthly_Achieved'] = np.where(fyp_within_month['Freelook'] >21, 
    fyp_within_month['FYP_Total'],0)
fyp_within_month = fyp_within_month.groupby(['Servicing Agent', 'FYP_month'])['FYP_Monthly_Achieved'].sum().reset_index()
fyp_within_month = fyp_within_month.pivot(index = 'Servicing Agent', columns = 'FYP_month', values='FYP_Monthly_Achieved').reset_index()
fyp_within_month[fyp_within_month['Servicing Agent'] == '60073218']
```

<h5>3.5 Tổng hợp lại thông tin của FYP

```python
fyp_month_achieved = fyp_cal.fillna(value=0).rename(columns={'FYP_Current(T)':'FYP_Achieved Monthly'})
fyp_month_achieved['FYP_Accum'] = fyp_month_achieved['FYP(T-2)'] + fyp_month_achieved['FYP(T-1)'] + fyp_month_achieved['FYP_Achieved Monthly']
fyp_month_achieved = fyp_month_achieved[['Servicing Agent', 'FYP_Achieved Monthly', 'FYP_Accum']]
#fyp_month_achieved chứa thông tin FYP trong tháng và FYP cộng dồn tới từ tháng T-2
fyp_ack = fyp_paid.fillna(value=0)
fyp_ack['FYP_Accum_ACK'] = fyp_ack['FYP_PCurrent(T)'] +  fyp_ack['FYP_P(T-1)'] +  fyp_ack['FYP_P(T-2)']
fyp_ack = fyp_ack[['Servicing Agent', 'FYP_Accum_ACK']]
```

<h3> 4. Tính CaseCount

<h5>4.1 Tính tổng số lượng các HD

```python
cc_agent = premium_table
cc_agent = cc_agent[['Policy No', 'PREM TRAN NAME', 'Applied Premium Date', 'POLICY ACKNOWLED', 'Issuing Agent', 'Freelook', 'Policy Status']].drop_duplicates()
cc_agent['PREM TRAN NAME'] = cc_agent['PREM TRAN NAME'].str.strip() #lưu ý làm sạch khoản trắng để có thể so sánh
cc_agent = cc_agent[(cc_agent['PREM TRAN NAME'] == 'Contract Issue') &
            ~cc_agent['Policy Status'].isin(['CF', 'DC', 'FL',  'NT', 'WD', 'PO'])]
cc_agent[cc_agent['Issuing Agent'] == '60073218']
```

<h5> 4.1 Lấy ra tổng số HD theo cột Applied (kể cả chưa có ACK)

```python
cc_cal = pd.merge(cc_agent, agent_target, left_on = 'Issuing Agent', right_on ='Agent_Number', how = 'left')
cc_cal = cc_cal[['Policy No', 'PREM TRAN NAME', 'Applied Premium Date',
       'POLICY ACKNOWLED', 'Issuing Agent', 'Date_Appointed' ,'Freelook', 'Policy Status', 'Date_On_Set', 'Date_1st_Temp', '1st_month', '2nd_month', '3rd_month']]
cc_cal['Applied_Temp'] = cc_cal['Applied Premium Date'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
cc_cal['ACK_Temp'] = cc_cal['POLICY ACKNOWLED'].dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
cc_cal = cc_cal[(cc_cal['Applied Premium Date'] >= cc_cal['Date_Appointed']) & (cc_cal['Applied Premium Date'] <= cc_cal['Date_On_Set'])]
cc_cal['CC_Month'] = np.where(cc_cal['Applied_Temp'] ==cc_cal['1st_month'], "CC(T-2)",
                    np.where(cc_cal['Applied_Temp'] == cc_cal['2nd_month'], "CC(T-1)",
                    np.where(cc_cal['Applied_Temp'] == cc_cal['3rd_month'], "CC_Current", "CC(T-2)")))
cc_app = cc_cal.groupby(['Issuing Agent', 'CC_Month'], dropna=False)['Policy No'].count().reset_index()
cc_app = cc_app.pivot(index='Issuing Agent', columns='CC_Month', values ='Policy No').reset_index()
cc_app[cc_app['Issuing Agent'] == '60073218']
```

<h5> 4.2 Lấy ra tổng số HD theo cột POLICY ACK DATE (Đây là số cộng dồn)

```python
cc_ack = cc_cal
cc_ack['CC_Month'] = np.where(cc_ack['ACK_Temp'] ==cc_ack['1st_month'], "CC_P(T-2)",
                    np.where(cc_ack['ACK_Temp'] == cc_ack['2nd_month'], "CC_P(T-1)",
                    np.where(cc_ack['ACK_Temp'] == cc_ack['3rd_month'], "CC_P_Current", "")))
cc_ack = cc_ack[cc_ack['CC_Month'] !=''] #LƯU Ý ko lấy các cột trống
cc_ack = cc_ack.groupby(['Issuing Agent', 'CC_Month'], dropna=False)['Policy No'].count().reset_index()
cc_ack = cc_ack.pivot(index='Issuing Agent', columns='CC_Month', values ='Policy No').reset_index()
cc_ack[cc_ack['Issuing Agent'] == '60073218']
```
