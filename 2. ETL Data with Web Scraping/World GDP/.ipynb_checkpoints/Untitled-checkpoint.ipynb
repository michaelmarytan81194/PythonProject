<h4> 1. Import necessary libraries <h4>

```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
#variables 
url = 'https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29'
table_attribs = ['Country','GDP_USD_millions']
```

<h4> 2. Extract functions <h4>

```python
page = requests.get(url).text #turn this into string
data = BeautifulSoup(page, 'html.parser')
df = pd.DataFrame(columns=table_attribs)
tables = data.find_all('tbody') #tbody means the tables in web content under BeautifulSoup instance
rows = tables[2].find_all('tr') #we find the 3rd tbody (table) is the one we need
for row in rows:
    col = row.find_all('td')
    if len(col) !=0:
        if col[0].find('a') is not None and '—' not in col[2]:
            data_dict ={"Country": col[0].a.contents[0],
            "GDP_USD_millions": col[2].contents[0]
            }
            df1 = pd.DataFrame(data_dict, index=[0])
            df = pd.concat([df, df1], ignore_index = True)
df
```
