```python
import pandas as pd
handle_excel = pd.ExcelFile('FM_2023.xlsx')
data = handle_excel.parse('Export')
data.head(10)
```

```python
period = data['Period'].unique()
period
```

```python
for term in period:
    df = data[data['Period'] == term]
    df.to_excel(f"{term}.xlsx", index = False)
```
